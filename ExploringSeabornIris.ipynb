# Step 3: Data Visualisation

- Pairplot
- FacetGrip

```python
sns.pairplot(data=iris, hue="species")
```

```python
g = sns.FacetGrid(iris, col="species", hue='species')
g.map(plt.scatter, "sepal_length", 'sepal_width')
```

```python
x =sns.FacetGrid(iris, col="species", hue="species")
x.map(plt.scatter, "petal_length", "petal_width")
```

From the data above we can see that both the sepal and petal size for **Virginica** are the broadest and biggest species followed by **Versicolor** species followed by **Setosa** species
