```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

sns.set_style('darkgrid')
```

# Learning Heatmap and Line plot using seaborn flight

```python
#Load the dataset

flight = sns.load_dataset('flights')
flight.head()
```

```python
#Pivot the dataset with `Year` as columns, `Months` as index, and `Passengers` as values.

flight_pivot = flight.pivot(index='month',columns='year',values='passengers')
flight_pivot.head()
```

# Heatmap

```python
#Adjusting the figsize
plt.figure(figsize=(10,8))

#Heatmap
sns.heatmap(flight_pivot,cmap="Blues")

#Label Details
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Flight Passengers')

plt.show()
```

```python
#Adjusting the figsize
plt.figure(figsize=(10,8))

#Heatmap
sns.heatmap(flight_pivot,cmap="YlGnBu",fmt="d",annot=True, linewidths=1)

#Label Details
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Flight Passengers')

plt.show()
```

```python
#Figure size
plt.figure(figsize=(8,6))

#Lineplot
sns.lineplot(x="month", y="passengers", data=flight)

#Label Details
plt.xlabel("Month")
plt.ylabel("Passengers")
plt.title("Flight Passengers Over Time")

plt.show()

#This show the level which passengers board Flights with respect to Months
```

```python
#Figure size
plt.figure(figsize=(10,8))

#Lineplot For Passengers over the Years
sns.lineplot(x="year", y="passengers", data=flight)

#Label Details
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.title("Flight Passengers Over Time")

plt.show()

#This shows the study rise of passengers boarding flights over time.
```
